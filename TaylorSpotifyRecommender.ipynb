```python
playlists = []
offset = 0
limit = 10
total_results = 0

# Perform an initial search to get the total number of results
result = sp.search("Taylor Swift", limit=10, offset=0, type="track", market="ES")
total_results = result["tracks"]["total"]

while offset < total_results:
    playlists.append(sp.search("Taylor Swift", limit, offset, "track", "ES"))
    offset += limit

result
```

```python
names = []
id = []
popularity  = []
release_date = []
total_tracks = []
albums = []

for i in range(len(split_df)):

    for j in range(len(split_df.columns)):

        names.append(split_df.iloc[i, j]['name'])
        id.append(split_df.iloc[i,j]['id'])
        popularity.append(split_df.iloc[i,j]['popularity'])
        release_date.append(split_df.iloc[i,j]['album']['release_date'])
        total_tracks.append(split_df.iloc[i,j]['album']['total_tracks'])
        albums.append(split_df.iloc[i,j]['album']['name'])
```

```python
list_of_tracks = pd.DataFrame({"Track_names": names, "id": id, "album_name": albums, "popularity" : popularity, "release_date": release_date, "total_tracks": total_tracks})
list_of_tracks
```

```python
len_of_tracks = len(list_of_tracks)
audio_fea = []
i = 0
while i < len_of_tracks:
    res = sp.audio_features(list_of_tracks['id'].iloc[i:i+100])
    audio_fea.extend(res)
    i += 100
```

```python
audio_features = pd.json_normalize(audio_fea)
audio_features.head()
```

```python
list_of_tracks_with_features =  pd.merge(list_of_tracks,audio_features)
```

```python
audio_features = audio_features.drop(["type","id","uri","track_href","analysis_url"], axis = 1)
```

```python
y_train = audio_features["energy"]
x_train = audio_features.drop("energy", axis = 1)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsRegressor
```

```python
def case_based_recommendations(query, n=10):
    scaler = MinMaxScaler()
    df_train_scaled = scaler.fit_transform(x_train)
    query_scaled = scaler.transform(query)
    knn = KNeighborsRegressor(n_neighbors=n + 1)
    knn.fit(df_train_scaled, y_train)
    distances, indices = knn.kneighbors(query_scaled.reshape(1, -1))

    top_song_indices = indices.flatten()[1:(n + 1)]  # Exclude the first nearest neighbor (query song)
    top_song_scores = distances.flatten()[1:(n + 1)]  # Exclude the first nearest neighbor (query song)

    top_songs = x_train.iloc[top_song_indices].copy()
    top_songs['Predicted Score'] = top_song_scores

    return top_songs
```

```python
top_songs =  pd.DataFrame(case_based_recommendations(query, n=10))
```
